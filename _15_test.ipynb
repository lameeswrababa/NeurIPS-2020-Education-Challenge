```python
sample_data_answer = sample_data.merge(answer_meta, how='left', on='AnswerId')
sample_data_answer.head()
```

```python
sample_data_answer_student = sample_data_answer.merge(student_meta, how='left', on='UserId')
sample_data_answer_student.head()
```

```python
from ast import literal_eval
question = []
#for index, row in df.iterrows():
for row in question_meta.itertuples(index=False):
  x = list (row)
  lis = list(literal_eval(x[1]))
  x.append (lis[-1] ) 
  a = x + lis
  question.append(a)

question = pd.DataFrame(question)
question.head()
```

```python
columns =['QuestionId','SubjectList','SubjectId','SubjectId_1','SubjectId_2','SubjectId_3','SubjectId_4','SubjectId_5','SubjectId_6','SubjectId_7','SubjectId_8','SubjectId_9','SubjectId_10','SubjectId_11','SubjectId_12','SubjectId_13','SubjectId_14']
question.columns = columns
question.head()
```

```python
sample_data_answer_student_question = sample_data_answer_student.merge(question, how='left', on='QuestionId')
sample_data_answer_student_question.head()
```

```python
sample_data_answer_student_question_subject = sample_data_answer_student_question.merge(subject_meta, how='left', on='SubjectId')
sample_data_answer_student_question_subject.head()
```

```python
data = pd.read_csv('/content/drive/MyDrive/DatasetProject1/Test_Project1.csv')
data.head()
```

```python
data = data.drop(['Unnamed: 0','Name','SubjectList','SchemeOfWorkId','Confidence','SubjectId_14','SubjectId_13','SubjectId_12','SubjectId_11','SubjectId_10','SubjectId_9','SubjectId_8','SubjectId_7','SubjectId_6','SubjectId_5'],axis=1)
```

```python
data['Gender']=data['Gender'].replace(3,0)
```

```python
list(data['Gender'].unique())
```

```python
data['DateOfBirth'] =pd.to_datetime(data['DateOfBirth'], errors = 'coerce')
data['DateAnswered'] =pd.to_datetime(data['DateAnswered'])
```

```python
data['Age'] = (data['DateAnswered'] - data['DateOfBirth']) / np.timedelta64(1,'Y')
```

```python
data.loc[data[data['Age'] < 6].index,'Age'] = 6
data.loc[data[data['Age'] > 45].index,'Age'] = 45
data['Age'] = data['Age'].fillna(data['Age'].mean())
```

```python
data['Age'] = data['Age'].astype(int)
```

```python
data = data.fillna({'ParentId':0})
data['ParentId'] = data['ParentId'].astype(int)
list(data['ParentId'].unique())
```

```python
data=data.merge(perForTest, on='UserId', how='left')
data.head()
```

```python
data.info()
```

```python
data = data.drop(['Unnamed: 0','DateAnswered','DateOfBirth','QuizId_x','QuizId_y','UserId'],axis=1)
```

```python
data.isnull().sum()
```

```python
data = data.fillna({'PremiumPupil':0})
data['PremiumPupil']=data['PremiumPupil'].replace(1,0)
data['PremiumPupil']=data['PremiumPupil'].replace(2,1)
```
