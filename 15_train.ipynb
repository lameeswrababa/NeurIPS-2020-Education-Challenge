```python
from ast import literal_eval
question = []
#for index, row in df.iterrows():
for row in question_meta.itertuples(index=False):
  x = list (row)
  lis = list(literal_eval(x[1]))
  x.append (lis[-1] ) 
  a = x + lis
  question.append(a)

question = pd.DataFrame(question)
question.head()
```

```python
data = pd.read_csv('/content/drive/MyDrive/DatasetProject1/toPer.csv')
data.head()
```

```python
data = data.drop(['Unnamed: 0','Name','SubjectList','SchemeOfWorkId','Confidence','SubjectId_14','SubjectId_13','SubjectId_12','SubjectId_11','SubjectId_10','SubjectId_9','SubjectId_8','SubjectId_7','SubjectId_6','SubjectId_5'],axis=1)
```

```python
data['Gender']=data['Gender'].replace(3,0)
```

```python
list(data['Gender'].unique())
```

```python
data['DateOfBirth'] =pd.to_datetime(data['DateOfBirth'], errors = 'coerce')
data['DateAnswered'] =pd.to_datetime(data['DateAnswered'])
```

```python
data['Age'] = (data['DateAnswered'] - data['DateOfBirth']) / np.timedelta64(1,'Y')
```

```python
data.loc[data[data['Age'] < 6].index,'Age'] = 6
data.loc[data[data['Age'] > 45].index,'Age'] = 45
data['Age'] = data['Age'].fillna(data['Age'].mean())
```

```python
data['Age'] = data['Age'].astype(int)
```

```python
data = data.fillna({'ParentId':0})
data['ParentId'] = data['ParentId'].astype(int)
list(data['ParentId'].unique())
```

```python
#student study before the exam
from typing import Counter
the_data = data.groupby(['UserId','QuizId']).apply(lambda group:(group['UserId'].mean(),group['QuizId'].mean(),(group['IsCorrect'].sum() / group.shape[0]))).reset_index(drop=True)
data_new = pd.DataFrame(the_data.tolist())
columns =['UserId','QuizId','ExamPreparation']
data_new.columns = columns
```

```python
data_new.head()
```

```python
data = data.merge(data_new, how='left', on=['UserId','QuizId'])
data.head()
```

```python
#overall student score in all questions he/she answed
overall = data.groupby(['UserId']).apply(lambda group:(group['UserId'].mean(),(group['IsCorrect'].sum() / group.shape[0]))).reset_index(drop=True)
overall_pd=pd.DataFrame(overall.tolist())
columns =['UserId','Student performance']
overall_pd.columns = columns
```

```python
overall_pd.head()
```

```python
data = data.merge(overall_pd, on='UserId', how='left')
data.head()
```
